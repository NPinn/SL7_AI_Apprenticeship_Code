# Euclidean Space

- Fundamental space of ***classical geometry***.
- The world we experience has $3$ dimensions
- But the ideas generalise to $N$ dimensions
- For more information, visit [Wikipedia Page](https://en.wikipedia.org/wiki/Euclidean_space)

Euclidean space was introduced by ancient Greeks as an abstraction of our physical space. This abstraction has three dimensions, but the same theories mean that Euclidean spaces of any non-negative integer dimension are possible (e.g. high dimensionality is common in Machine Learning). 

One way to think of the Euclidean plane is as a set of points satisfying certain relationships, expressible in terms of distance and angles. In a Euclidean space, there is a is a clear notion of location and distance between any different chosen points in the space.

# Euclidean Space



The figure shows an *Euclidean space* of 3 dimensions \\(X, Y\\) and \\(Z\\). The dimensions are perpendicular, or orthogonal, to each other (\\(X \perp Y \perp Z\\)). A point (\\(x, y, z)\\) in this space can be represented by its distance from origin \\(O\\) in each dimension (length of pink lines in the figure). That is, the point (\\(x, y, z)\\) is \\(x\\) units away from \\(O\\) in \\(X\\) dimension, \\(y\\) units away from \\(O\\) in \\(Y\\) dimension and \\(z\\) units away from \\(O\\) in \\(Z\\) dimension where \\(x, y, z \in \mathbb{R}\\) and \\(x, y, z \in \{-\infty, +\infty\}\\) 

In linear algebra, a point in an $n$-dimensional Euclidean space can be represented with $\mathbb{R}^n$, a vector of \\(n\\) values (e.g. point ($x, y, z$) in 3-dimensional Euclidean space in the figure). The distance between two points in this vector space is represented by the scalar obtained by calculating the ***inner*** product between the vectors of the two points.  

A majority of the representations used in conventional machine learning can be conveniently expressed as part of Euclidean spaces. This allows us to use the powerful tools available in Euclidean geometry (distance, similarity etc.) to design pattern recognition algorithms. 

# Non-Euclidean Spaces

- There are many complex dimensional spaces
- Proximity is not as straightforward


The world we live in doesn't wholly comprise of data that fit in Euclidean spaces. There are many instances where representing the data in a non-Euclidean space would make much more sense. 

In a Euclidean space, knowing the location in terms of the coordinate system itself will allow us to easily compute the distance between different points in the space (take the inner products of the $\mathbb{R}^n$ vector). However, there are numerous scenarios where the data may come from a different coordinate system, where the rules of Euclidean spaces might not apply.

- It is not possible to roll the Earth out flat without introducing distortions (formally, a shape-preserving and area-preserving projection is not possible).

- Although the Earth itself lives in (is "embedded" within) 3D Euclidean space, the 2D surface of the Earth is non-Euclidean.



The most familiar example of a non-Euclidean space is the surface of the Earth.

It is well-known from cartography that it is not possible to roll the Earth out flat without introducing distortions (formally, a shape-preserving and area-preserving projection is not possible).

Although the Earth itself lives in (is *embedded* within) 3D Euclidean space, the 2D surface of the Earth is non-Euclidean.

# Example: Surfaces

- Consider 3D doughnut or torus
- Again, the surface of the doughnut/torus is non-Euclidean
- Shortest distances (geodesics) in the space are *not* 3D Euclidean distances



The second example we are going to look at is a 3D doughnut surface. Although this example might look like it can very easily fit in a Euclidean space, we need to look more carefully.

It can definitely be represented in a 3D space where every point in the surface has a value in the 3D Euclidean space. But the proposed 3D Euclidean space will also include a large number of locations (points) that do not belong to the doughnut (the empty locations), which might be a problem for us. 

To identify the limitations of using a Euclidean space for representing the doughnut, let us use a real-world scenario. Imagine that an ant found its way on to the surface of this doughnut which is suspended in mid-air. If the ant landed on point ***A*** and wants to get to point ***B***, what would be the distance the ant has to travel? Given that the doughnut is  ***suspended in mid air***, the green line in the figure represents the *correct distance*, rather than the red line which will be the distance representation in the Euclidean space.

Sometimes, we come across machine learning tasks that might demand representations that behave in non-continuous spaces and might not fit in Euclidean spaces that we are comfortable with.

# Example: Hierarchy



The next example that we can look at is a hierarchy. A hierarchy contains a set of points that are connected to each other in a tree-like manner. Firstly, there is no clear notion of location for the points. In a Euclidean space, we work with a coordinate space where every point has a specific location that can be represented using a vector in the dimension space. In this example, the location of each point in the hierarchy is relational to the other points. Therefore the location is proximal rather than static. 

We also observe the non-continual nature in this example. There is no necessity for a distance to exist between every pair of points in the space. This requires us to look beyond Euclidean spaces to find more powerful representations.

# Graph

- Mathematical structure used in representing ***pairwise relations between objects***
- Consists of ***vertices*** and ***edges***
    


A graph is a data structure that allows us to embed entities, and their relationships to other entities, in a compact representation. A graph is a more abstract version of a hierarchy, as hierarchies belongs to the "tree" data structure category (which is a subset of graph data structures). Graphs allow us to embed pairwise interactions between objects.

The figure shows a graph of 5 objects (vertices), and the relationships between these 5 objects (edges). Notice that graphs carry many qualities that we observed with hierarchies (because hierarchies are a subgroup of graphs), and might be better represented by an alternative space rather than a Euclidean space.

# Vertices and Edges




- Vertices (nodes): represents the finite set of objects of interest
- Edges (links): represents a finite set of unordered connections between pairs of vertices
\\[\mathcal{G} = (\mathcal{V}, \mathcal{E})\\]
    

A graph has two main components.
1. Vertices
2. Edges

***Vertices*** in a graph represent a finite number of objects of interest. In the figure, the vertices are represented with pink circles. Vertices are also identified as "nodes" in a graph. 

Vertices in a graph are connected to each other using ***edges***. These represent the presence of an interaction (or absence of one) between a pair of vertices in the graph. They are also identified as "links" in the graph. The blue lines that connect a subset of pairs of nodes in the above figure are the edges of the graph. 

Vertices and edges together express a graph.

# Types of Graphs

- Undirected vs. Directed
- Uniform vs. Weighted



When learning about graphs, one should learn the different types of graphs that exist. There are 4 main different types of graphs that are noteworthy in terms of the different types of information they embed.

*Undirected vs. directed* graphs distinguish if the graph representation provides information about the direction of the edges. The left sub-figure illustrates this difference. Although an edge in a graph represents a relationship between two nodes (left, left), sometimes, this relationship is directional (left, right). In a scenario where all the connections are bidirectional (e.g. friends in a social network), an undirected graph is sufficient. In cases where the connections can be unidirectional (e.g. hyperlinks), a directed graph is required. 

*Unweighted vs. weighted* graphs differ on whether the magnitude of the relationship can be represented (two rightmost graphs). In some situations, treating all relationships equally makes sense. In this case, the weight of the edge between any given pair of connected nodes is the same and an unweighted graph representation can be used. But sometimes, the degree of the relationship varies between different pairs. For example, we may wish to represent the bandwidth between two IoT devices. In this situation the weight of each edge will be different and a weighted graph representation is required.

# Types of Graphs

- Homogeneous vs. Heterogeneous
- Acyclic vs. Cyclic



The remaining 2 types of graphs are described here. 

*Homogeneous vs. heterogeneous* distinction comes into play when you have multiple different types of vertices and edges in a graph. A homogeneous (left, left) graph only has one node type and one edge type (e.g. social network between friends). A heterogeneous graph can have different types of nodes and different types of edges (e.g. a knowledge graph with node types such as 'Person', 'Location', etc. and edge types such as 'Part of', 'Parent of', etc.).

*Acyclic vs. cyclic* graphs distinguish themselves based on the presence of cycles in the graph. An acyclic graph (right, left) doesn't contain any cycles (e.g. a family tree) where as the presence of cycles can be observed in a cyclic (right, right) graph (e.g. link graph of the internet).

# Representing Graphs

- To do machine learning, we need to numerically represent graphs
- ***Adjacency Matrix*** and ***Degree Matrix*** can be used
- Adjacency Matrix \\({A}(\mathcal{G})\\) represents the connections in the graph
- \\(A(\mathcal{G})\\) is a ***square matrix***  where \\(dim(A)\\) is \\(|V| \times |V|\\)




To leverage the power of graphs in machine learning algorithms, one must be able to represent them numerically. The adjacency matrix and the degree matrix are two matrix representations that can represent the connectivity and degree of a graph in matrix form. These representations allow us to plug graph structures into linear algebraic expressions that carry out pattern recognition. 

An adjacency matrix represents the pairwise relationships in a graph. It is a square matrix where the elements represent (1) presence of an edge between two nodes and (2) the magnitude of the connection. Adjacency matrix $A(\mathcal{G})$ of graph $\mathcal{G}$ is a ***square matrix***  of shape $|V| \times |V|$ where $|V|$ represents the number of *vertices* in the graph. The position $(a,b)$, where $a,b \in V$, gives information about the degree of connectivity between vertices $a$ and $b$.

Adjacency matrices can be built for all sorts of graphs and mainly allow representing the magnitude and direction of a relationship.

# Learning Tasks Associated with GNNs

- There are ***three*** main classes of learning tasks:

1. Node level
2. Link level
3. Graph level

# Node-level Labeling

- Learns an ***embedding for nodes*** in the graph
- The representation is used to do predictions based on node (Eg. node classification)

# Link-level Labeling

- Related to ***link prediction*** and ***edge classification*** tasks
- Learns a neural network that can predict the type of connection or strength of an edge

# Graph-level Labeling
- Learns a representation that represents the ***whole graph***.
- In supervised learning, every observation is a `(graph, label)`

# Foundations of GNN

## Problem Setting

- $\mathcal{V}$ is the ***vertex set*** of the graph
- $\mathcal{A}$ is the ***adjacency matrix***
- ${X} \in \mathbb{R}^{d \times |\mathcal{V}|} $ is a matrix of ***node features***
    - Social network: User profile
    - Recommendation system: Item metadata
- Sometimes ***no*** node features
    - Indicator variables/one-hot encoding
    - Vector of constants (e.g. [1, 1, ... 1])

# Node Representation
- Every node constructs a local computation graph of its neighbours
- Aggregate neighbour information using a ***neural network*** (white cog-wheels)



# Aggregation 

$$
h_A =  \color{blue}{f \Big(} \color{green}{\sum_{u \in N(A)} \frac{h_{u}}{|N(A)|}} \color{blue}{\Big)},
$$

where
- $\color{green}{\sum_{u \in N(A)} \frac{h_{u}}{|N(A)|}}$ represents ***average embedding*** for neighbours of node $A$, represented by the green circle in the figure $( N(A) \in \{B, C, D\}$).
- $\color{blue}{f(\cdot)}$ is the ***neural layer*** that transforms the average embedding, corresponding to the blue-circled cog-wheel in the figure.



# Depth (# Layers)

- The model can have ***arbitrary*** number of layers
- $n^{th}$ layer gets information from nodes that are $n$ hops away 
    - Less layers: immediate neighhbours
    - More layers: further nodes
- Deeper networks lead to over-smoothing



# Neighbour Averaging

$$\color{green}{h_v^0} = x_v$$
$$
h_v^k = \color{orange}{\sigma} \left( W_k \color{blue}{\sum_{u \in N(v)} \frac{h_{u}^{k-1}}{|N(v)|}} + B_k\color{red}{h_v^{k-1}} \right), \forall k \in \{1,\dots, K\}
$$
$$\color{pink}{z_v} = h_v^K$$

where:
- $\color{green}{h_v^0}$ is the layer 0 embedding equivalent to node features
- $\color{orange}{\sigma}$ is the non-linear transformation (e.g. ReLU)
- $\color{blue}{\sum_{u \in N(v)} \frac{h_{u}^{k-1}}{|N(v)|}}$ represents the average embedding for neighbours of $v$
- $\color{red}{h_v^{k-1}}$ is the previous layer embedding of $v$
- $\color{pink}{z_v}$ is the final embedding of node $v$

# Model Parameters

$$h_v^0 = x_v$$
$$
h_v^k = \sigma \left( \color{orange}{W_k} \sum_{u \in N(v)} \frac{h_{u}^{k-1}}{|N(v)|} + \color{orange}{B_k}h_v^{k-1} \right), \forall k \in \{1,\dots, K\}
$$
$$z_v = h_v^K$$

- The weights matrices \\(\color{orange}{W_k}\\) and \\(\color{orange}{B_k}\\) are trainable.
- Can use ***stochastic gradient descent*** to learn weight parameters.

# Vector Form

$$h_v^0 = x_v$$
$$
h_v^k = \sigma \left(  \color{orange}{W_k} \sum_{u \in N(v)} \frac{h_{u}^{k-1}}{|N(v)|} + \color{orange}{B_k}h_v^{k-1} \right), \forall k \in \{1,\dots, K\}
$$
$$z_v = h_v^K$$

Can be written in ***vectorised form*** as:

$$H^{k} = \sigma \left(\color{green}{\mathcal{A}} H^{k-1} \color{orange}{W_{k}} + H^{k-1} \color{orange}{B_{k}} \right)$$
$$where \texttt{ } \color{green}{\mathcal{A}} = \mathbf{D}^{-\frac{1}{2}}\mathbf{A}\mathbf{D}^{-\frac{1}{2}}$$

- $\color{black}{\mathbf{A}}$ is the ***adjacency matrix*** 
- $\mathbf{D}$ is the ***degree matrix***

# Loss Function

- A loss function can be used to direct the training process
- Could use ***unsupervised approach***
    - graph structure only
    - similar nodes have similar embeddings
    - e.g. node2vec and DeepWalk

# Supervised GNN
- Could be trained using labels in supervised manner
- Loss incorporates the labels
- e.g. Node classification
$$\mathcal{L} = \sum_{v \in V} \color{orange}{y_v} \log(\sigma (\color{blue}{z_v^\intercal} \color{red}{\theta})) + (1-\color{orange}{y_v}) \log(1-\sigma (\color{blue}{z_v^\intercal} \color{red}{\theta}))$$
- $\mathcal{L}$ is the loss
- $\color{orange}{y_v}$ is the actual label
- $\sigma (\color{blue}{z_v^\intercal} \color{red}{\theta})$ is the prediction where $\color{blue}{z_v^\intercal}$ is the learned node embedding of node $v$, and $\color{red}{\theta}$ represents the learned weights (feed-forward layer) for classifying nodes.

# Neighbour Averaging

$$h_v^0 = x_v$$
$$h_k^v = \sigma \left( W_k \sum_{u \in N(v)} \frac{h_{u}^{k-1}}{|N(v)|} + B_kh_v^{k-1} \right), \forall k \in \{1,\dots, K\}
$$
$$z_v = h_v^K$$

# Deep Graph Library (DGL)

- Doesn't extend a specific framework
- Backend-agnostic [networkX](https://networkx.org/)-like API
- PyTorch, TensorFlow, and MxNet backends compatible 
- Keras for geometric deep learning
- Development led by NYU and AWS
- Extensive documentation available [here](https://www.dgl.ai/)



# Spektral

- Another backend-agnostic API for graph deep learning

- Development led by Daniele Grattarola at Università della Svizzera Italiana (CH)
- Very easy to learn and prototype
- Extensive documentation available [here](https://graphneural.network/)

# Taxonomy of GNNs

- Using the nodes in the neighbourhood in the graph to learn better representations 
- ***Four*** main types of GNNs
- Convolutional GNNs (ConvGNN)
    - Generalises ***convolution*** from grid data to graph data
- Recurrent GNNs (RecGNN)
    - Using ***RNN*** architectures to exchange information with neighbours

# Taxonomy of GNNs

- Graph Autoencoders (GAE)
    - ***Unsupervised*** approach to learning network embeddings that can reconstruct the graph
- Spatial-temporal GNNs (STGNN)
    - Increasingly useful in advanced tasks that have both ***space-time*** dimensions
    - E.g: Traffic forecasting, human action recognition, etc.
