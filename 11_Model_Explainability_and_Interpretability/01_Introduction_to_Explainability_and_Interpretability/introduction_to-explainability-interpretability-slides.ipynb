## Black Box Models

- Deep Models can have ***millions to billions of parameters*** 
- Complexity gives ***extraordinary accuracy***

When talking about machine learning models that perform well, one cannot escape getting into the high complexity of highly performant machine learning models. These highly complex models are defined as "Black Box" models. 

Most of the recent breakthroughs of Artificial Intelligence and Machine Learning has happened through highly complex models that come from families of models such as ensembles and deep learning. For example, ensemble models such as Random Forest model creates 100s or 1000s of decision trees that make the final prediction "together". The complex structure of these models make them very challenging for human intuition. Similarly, recent deep learning models usually train millions or billions of parameters making them impossible for a human expert to understand. 

The common factor between such complex models is that they usually lead to very high accuracy in terms of making predictions (extraordinary accuracy). However, this performance comes at the cost of transparency and understandability of these models. The challenge then is how can we trust the decisions these models make when we have no easy to way to understand how they came to such a conclusion.

## Some models are ***easy*** to interpret

- Linear/Logistic Regression
    - Weights for individual features
    - Presents exact contribution to prediction
        - $y = 3.5 x_1 + 4.21 x_2$
        - increasing ***x1*** will increase y by 3.5 units

Let's look at white box models that are fairly simple to understand. Here we have taken linear regression and logistic regression, two popular models that are used to solve regression and classification problems respectively. Both these models learn a system of weights that correspond to every every input feature. In linear regression, these weights correspond to how the label changes when the input feature changes by one unit. In logistic regression, it corresponds to how the odds of the output label changes when the input changes. 

From the definitions themselves, we can observe that the internal weights of these two models are intuitive, enabling human experts to understand these models with minimal effort.  

## Some models are ***easy*** to interpret

## Some models are ***harder*** to interpret



- Ensemble models (Random Forest, Gradient Boosting Machines, etc…)
    - Hard to understand the role of each feature
    - Need to calculate ***feature importance***




From decision trees, we step into ensembles of decision trees that make predictions together. In contrast to a simple decision tree model, the model has become more complex now. The figure shows that there are `n` decision trees in this Random Forest model which will use the rules learned in individual trees to make an prediction, leading to `n` predictions. Then an aggregation function  considers all these predictions to decide what the ultimate label should be. 

The model is no longer simple. The link between input features and the output labels is now fairly complex due to multiple rules within multiple trees, leading to multiple predictions which are then aggregated into a single decision. With very few additions to the Decision Tree model, we have created the Random Forest model which is significantly more complex, leading to further difficulties in interpreting the model. 

Due to the high complexity of the model, post-hoc techniques such as feature importance analysis should be used to understand how different input features influence the final labels.

## Interpretability Vs. Explainability

- Interpretable Model
    - A model that ***can be directly explained and interpreted*** by human experts
    - Trade off Interpretability vs Performance (due to ***limited complexity***)
- Explainable Model
    - A model that is ***too complex*** for a human to understand
    - ***Additonal tools/ methods/ techniques*** need to be devised to "explain" how the model works

We have now discussed how increasing the complexity of machine learning models makes it increasingly hard for humans to intuitively understand their inner-workings. This connection between complexity and intuitive understanding allows us to define interpretable models and explainable models. 

As the definition goes, an interpretable model is a model that allows a human expert to directly understand what patterns are used by the model to make predictions. Easy interpretation is possible in interpretable models due to their lack of complexity which comes at the price of inferior predictive performance. 

On the contrary, an explainable model is a complex machine learning model where direct human "interpretation" of the model is unrealistic. Due to this, additional tools and techniques are needed to create "explanations" of how the model operates. Due to this quality, these models are defined as explainable models. Due to their higher complexity, explainable models usually demonstrate superior predictive performance over interpretable models.

## Interpretability Vs. Explainability



Model interpretability and explainability are intrinsically linked to model complexity. As demonstrated in the figure, less complex methods such as decision trees, linear and logistic regression may be consider "interpretable". This is because they are "white box" models which can be directly understood by human experts.

At the other end of the complexity spectrum, we find "explainable" models. These are highly complex, black box models which do not permit direct understanding. In these circumstances, additional tools are needed to extract explanations. These models include random forests, feed-forward neural networks, convolutional neural networks, recurrent neural networks, long short-term memory networks, to name a few.

## Explainability Accross the ML Pipeline

- Pre-modelling
    - Analysing the ***dataset***: characteristics of the dataset, class imbalances, feature distributions, biases etc.
    - Analysing ***features***: important features, sparse features, correlations etc.
    - Choice of ***humanly intuitive*** features

Different techniques should be used in different parts of the machine learning pipeline to gain humanly intuitive insights. Some explanations should be obtained in the pre-modelling phase of the pipeline where feature engineering happens. 

One aspect in the pre-modelling phase is to identify the characteristics of the dataset at hand. In this phase, different descriptive analyses such as frequency distribution analysis and correlation analysis can be done to identify the common patterns that are present in the dataset. This also allows the data scientist to understand the characteristics of different features.

One important point to present here is that choosing humanly intuitive features as input to your machine learning model is important in making the subsequent model explanations meaningful. The majority of model explanation techniques detect and present the patterns that map the inputs to the outputs in a humanly intuitive fashion. These mappings will not be meaningful if the input features we choose are too complex for human understanding. Such a scenario will make the explanations useless. 

### Towards Human-Intuitive Explainations
- ***Different Explanations*** for Text, Tabular and Image Data
- Explanations through ***Examples***



Assuming that intuitive input features are chosen, there are different model explanation techniques for different types of data. Some of these examples are presented in the figure presented here. 

When looking at tabular data, explanations usually consist of information such as feature importance which identify what columns in the table are most important for the model. Similarly, for text data, the explanations usually map input word tokens to output labels. For image data, the mapping happens between the input pixels in the image and the output labels.

From the pre-modelling stage, we go to the actual model training phase. The first decision one can make here is between what type of a machine learning model is utilised for the application. We can choose between a white box model or a black box model. White box models are inherently simple models that enable simple model interpretation. Choosing a white box model will eliminate the need to devise additional techniques to understand/explain the model. However, this comes at a cost. That cost is reduced predictive performance. Simple models have less capacity to identify patterns, making them weaker learners compared to their black box counterparts.

Therefore, the choice of model white box vs. black box itself is one aspect of model explanation. If a white box model is chosen, the model internals (such as weights, rules etc.) are easily accessible making them intrisically interpretable. However, choosing a black box model (which is likely to improve predictive performance) will require post-hoc explanation tools.

## Taxonomy of Model Interpretability and Explainability

- The literature surrounding interpretability and explainability encompasses multiple taxonomies
- The knowledge areas can be viewed from multiple dimensions
- Main separation: ***Interpretability vs. Explainability***
- ***Multiple criteria*** can be used for further differentiation of the taxonomy.

We established at the beginning that Explainability and Interpretability relate to the tools and techniques used to bridge that gap between expert understanding and internal model mechanics.

The main categorisation lies between interpretable (white box) models  and explainable (black box) models. Beyond that, methods for machine learning interpretability can be classified according to various criteria.

### Intrinsic Vs. Post-Hoc 

This criterion distinguishes whether interpretability is achieved by restricting the complexity of the machine learning model (intrinsic) or by applying methods that analyse the model after training (post hoc).

- ***Intrinsic Approaches*** achieve model explainability by using less complex models
    - Uses ***White Box*** models
    - The Models are ***Interpretable***
- ***Post-Hoc Approaches*** are used when the inherent complexity of model prohibits interpretation
    - Used with ***Black Box*** models
    - ***Additional methods/ tools*** are used to achieve interpretability
    - The models are ***Explained***

Another important criterion in the interpretability vs explainability categorisation is whether transparency is intrinsic to the model, or if it is achieved through additional, post-hoc explanation methods.

In general, interpretable models are those which are less complex and can be directly understood through a survey of their coefficients/rules. Intrinsic interpretability is achieved by utilising white box models. 

On the contrary, black box models have complex structures (such as ensembles, deep neural networks) and require additional methods to explain their predictions. This category is called post-hoc explainability approaches. 

### Model-specific Vs. Model-agnostic

- ***Model-specific***: Interpretation tools that are limited to specific models  
    - By definition, interpretaion of interpretable models is always model-specific
        - eg: Weights of a linear regression model
    - Interpretation techniques that are specific to a model family (even when model family is explainable)
        - eg: DeepLIFT for Deep Neural Networks
- ***Model-agnostic***: Interpretation tools that can be used to explain any machine learning model
    - By definition, cannot have access to model parameters (as they are model-specific)
    - Works by analysing input &rarr; output pairs
        - eg: [LIME](https://github.com/marcotcr/lime), [SHAP](https://github.com/slundberg/shap)

An alternative criterion to look at explainability approaches is through the lens of model-specific vs. model-agnostic methods.

Model-specific interpretation tools are restricted to specific model classes. What this means is that the interpretation technique only applies to very specific model family and doesn't generalise across all models. These model-specific approaches exist in both interpretable and explainable model domains. By definition, the interpretation of interpretable (white box) models fall under the model-specific category. One example for this is inspecting the weights of a linear regression. To understand why this is model-specific, one need only consider decision trees models, where inspecting the weights wouldn't work since the model does not contain weights.

However, even when the models are complex, some explanation techniques are model specific. For example, DeepLIFT (Deep Learning Important FeaTures) is a method for interpreting feature influence but which only applies to deep neural networks. This is because the method uses differences of "activations" that are only present in deep neural networks. 

Model agnostic interpretation tools are tools that can help understanding any model regardless of its model family. By definition they do not use model parameters as model parameters are model specific internals (e.g. weights in linear models vs. rules in decision trees). LIME is a good example of a model-agnostic explainability technique as it builds a linear model using the features and predictions from the model. This way, the tool only relies on features (coming from the dataset) and predictions that can be obtained from any model, regardless of model family. This allows LIME to be model agnostic.

### Local Vs. Global Explanations

- ***Local***: Methods and techniques that explain individual predictions
    - Important for ***case-by-case*** analysis
    - Looks at the local prediction landscape
    - E.g. Loan applications 
- ***Global***: Methods that explain the behaviour of the entire model
    - Important for ***understanding the whole model***
    - Looks at multiple predictions
    - E.g. House price prediction, sentiment analysis

The other criterion one can use to distinguish between explanation approaches is global vs. local explanations. 

Local explanations focus on individual predictions. These approaches take individual predictions and attempt to explain those predictions in their local context. This type of explanations are very useful when case by case explanations are required. Some examples are loan application outcomes or disease diagnoses where rationalisations for individual cases is necessary.  

On the contrary, global explanation approaches attempt to analyse the entire model's behaviour. They attempt to look at the global behaviour of models across all possible predictions. This type of model is useful when future predictions will be done on observations that may not be captured by your training data. A good example is a sentiment analysis model where new text extracts need to be classified as positive or negative sentiment. In a situation like this, global model behaviour informs the stakeholders if the model is going to behave appropriately with a large quantity of future predictions.
